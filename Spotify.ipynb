#  Number of Releases in each month

# Correlation Heatmap: Audio Features

```python
features = df[['bpm', 'danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']]
cm = features.corr()
plt.figure(figsize=(8,4))
sns.heatmap(cm,cmap='crest',annot=True)
plt.title('Track Features Correlations', fontsize=12, fontweight='bold')
plt.show()
```

# Top 10 Artists Based on Total Streams

```python
df['streams']=pd.to_numeric(df['streams'],errors='coerce')
# Grouping by artists again and summing up their streams
artist_streams = df.groupby('artist(s)_name')['streams'].sum().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=artist_streams.values, y=artist_streams.index, palette="viridis")
plt.title('Top 10 Artists Based on Total Streams', fontsize=16,fontweight='bold')
plt.xlabel('Total Streams (in billions)',fontsize=12,fontweight='bold')
plt.ylabel('Artists Name',fontsize=12,fontweight='bold')
plt.show()
```

# Finding Underrated Songs on Spotify

```python
# Filter songs with high audio features
high_danceability = df['danceability_%'] > df['danceability_%'].mean()
high_energy = df['energy_%'] > df['energy_%'].mean()
high_valence = df['valence_%'] > df['valence_%'].mean()
```

```python
# Combine filters with logical AND
high_audio_features = high_danceability & high_energy & high_valence
```

```python
# Filter songs with relatively lower streaming counts
low_streams = df['streams'] < df['streams'].median()
```

```python
# Combine filters with logical AND
underrated_songs = high_audio_features & low_streams
```

```python
# Get the underrated songs
underrated_songs_data = df [underrated_songs]
```

```python
# Display the underrated songs
print("Underrated Songs:")
print(underrated_songs_data[['danceability_%', 'energy_%', 'valence_%', 'streams']])
```

# Is there a relationship between energy levels and acousticness? 

```python
spotify_data_filtered = df.dropna(subset=['energy_%', 'acousticness_%'])
correlation_data = spotify_data_filtered[['energy_%', 'acousticness_%']]

# Calculate the correlation matrix
correlation_matrix = correlation_data.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(5,5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Between Energy Levels and Acousticness')
plt.show()
```

# Number of tracks released each month in 2023 varied over time

```python
monthly_track_counts = df.groupby(['released_year', 'released_month'])['track_name'].count().reset_index()
plt.figure(figsize=(8, 6))
plt.plot(monthly_track_counts['released_month'], monthly_track_counts['track_name'], marker='o', linestyle='-')
plt.title('Number of Tracks Released Each Month in 2023')
plt.xlabel('Month')
plt.ylabel('Number of Tracks')
plt.grid(True)
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.show()
```
